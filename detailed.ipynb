```python
amenities_distinct_values = {}

for i in range(len(df)):
  list_values = eval(df['amenities'][i])
  for value in list_values:
    if value not in amenities_distinct_values.keys():
      amenities_distinct_values.update({value: 1})
    else:
      amenities_distinct_values[value] += 1

amenities_distinct_values = {k: v for k, v in sorted(amenities_distinct_values.items(), key=lambda item: item[1], reverse=True)}
```

```python
amenities_distinct_values
```

```python
import ast
from sklearn.preprocessing import MultiLabelBinarizer

df['host_verifications'] = df['host_verifications'].apply(ast.literal_eval)

mlb = MultiLabelBinarizer()
verification_ohe = pd.DataFrame(mlb.fit_transform(df['host_verifications']),columns=[f"host_verification_{cls}" for cls in mlb.classes_], index=df.index)

df = pd.concat([df, verification_ohe], axis=1)

df.drop('host_verifications', axis=1, inplace=True)
```

```python
host_verifications_distinct_values = []

for i in range(len(df)):
  list_values = eval(df['host_verifications'][i])
  for value in list_values:
    if value not in host_verifications_distinct_values:
      host_verifications_distinct_values.append(value)

print(host_verifications_distinct_values)
```
